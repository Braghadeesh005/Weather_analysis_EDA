# Data Cleaning

```python
df.isnull()
```

Checking for the null values using Heatmap.

```python
sns.heatmap(df.isnull())
```

Dropping the features Hmdx and Wind Chill. Since it have very high null values.

```python
df.drop(['Hmdx','Wind Chill'],axis=1,inplace=True)
```

```python
sns.heatmap(df.isnull())
```

'Wind Dir (10s deg)' Feature have some null values which is to be filled.

```python
sns.boxplot(x='Month',y='Wind Dir (10s deg)',data=df)
```

From this boxplot, median of wind dir for each month can be found.

Creating a function to return the value of wind direction according the month

```python
def wind_direction(col):
    wind = col[0];
    month = col[1];
    
    if pd.isnull(wind):
        
        if month==1:
            return 11
        elif month==2:
            return 12
        elif month==3:
            return 14
        elif month==4:
            return 16
        elif month==5:
            return 18
        elif month==6:
            return 16
        elif month==7:
            return 17
        elif month==8:
            return 13
        elif month==9:
            return 21
        elif month==10:
            return 12
        elif month==11:
            return 11
        elif month==12:
            return 10
            
    else:
        return wind
    
```

Applying the function to fill all the null values with approximate values.

```python
df['Wind Dir (10s deg)'] = df[['Wind Dir (10s deg)','Month']].apply(wind_direction,axis=1)
```

```python
sns.heatmap(df.isnull())
```

Now the heatmap shows no null values

```python
df.head()
```

Dropping Date/Time Feature as it is already available indivually as Year , Month , Day , Time

```python
df.drop(['Date/Time'],axis=1,inplace=True)
```
